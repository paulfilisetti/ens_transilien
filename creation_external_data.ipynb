```python
# formatage de la date de mon df

df['date'] = pd.to_datetime(df['AAAAMMJJ'].astype(str), format='%Y%m%d')

df.head()
```

```python
# là je suis clean normalement je drop la colonne date initiale
df = df.drop(columns=['AAAAMMJJ'])
df.head()
```

```python
# je veux juste mettre les vacances scolaires pour avoir une idée de la demande, et notamment de l'été avec travaux


# Définition des vacances scolaires 2023 - Zone C
vacances = [
    ("Hiver", "2023-02-18", "2023-03-06"),
    ("Printemps", "2023-04-22", "2023-05-09"),
    ("Été", "2023-07-08", "2023-09-03"),
    ("Toussaint", "2023-10-21", "2023-11-06"),
    ("Noël", "2023-12-23", "2024-01-08")
]

# Fonction pour vérifier si une date est dans une période de vacances
def is_vacation(date):
    for _, start, end in vacances:
        if pd.to_datetime(start) <= date <= pd.to_datetime(end):
            return 1
    return 0

# Appliquer la fonction
df['Vacances_Scolaires'] = df['date'].apply(is_vacation)

df.head()
```

```python
# je renomme et regarde la tail pour voir si effectivement les vacances sont flag
df['vacances'] = df['Vacances_Scolaires']
df.tail()
```

```python
# je drop la colonne vacances_scolaire toute moche et ce sera bon
df = df.drop(columns=['Vacances_Scolaires'])
df.head()
```

# merge de mes df

```python
# je vais merge df_train et df. Ensuite pareil pour df_test
df_train = df_train.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
df_train.head()
```

```python
# j'applique ma fonction définit précédemment'

def format_df(df):
    # copy du df pour rien âbimer
    df = df.copy()

    # formatage des colonnes floats en int
    cols_to_convert = ['p2q0', 'p3q0', 'p4q0', 'p0q2', 'p0q3', 'p0q4']
    df[cols_to_convert] = df[cols_to_convert].astype(int)

    # conversion date en datetime
    df.date = pd.to_datetime(df.date)
    # faire ressortir les infos annee, mois, jour, weekend
    df['annee'] = df['date'].dt.year
    df['mois'] = df['date'].dt.month
    df['joursemaine'] = df['date'].dt.weekday
    # on enlève la colonne date maintenant
    #df = df.drop(columns=['date']) # je drop pas encore la date car je veux merge dessus
    return df

df_train = format_df(df_train)

df_train.head()
```
